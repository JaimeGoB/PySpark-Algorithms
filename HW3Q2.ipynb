Importing Libraries needed for PySpark

```
from pyspark import SparkContext
from pyspark import SparkConf
import numpy as np
```

## (a) Create a new **SparkContext**. Note that if you are using python, you will need to import SparkContext from pyspark.

```
sc = SparkContext.getOrCreate()
```

## (b) Create a RDD (with parallelize) with a sample of 500 integers from uniform distribution. Integers should go from 0 to 50 and for that you can use random.randint library.

```
rdd = sc.parallelize(np.random.randint(51, size=500))
```

## (c) Using Spark operations on RDD (transformation, actions or persistence), compute and print the following measures over the RDD created on item (b): sum, count and mean. Note that mean should be around 25.

#### SUM

```
sum = rdd.sum()
sum
```

#### COUNT

```
count = rdd.count()
count
```

#### MEAN

```
mean = rdd.mean()
```

## (d) Using spark operations on the same RDD (transformation, actions or persistence) compute and print the variance (it should be around 216.67).

#### Variance

```
variance = rdd.variance()
variance
```

## (e) Using Spark operations on RDD (transformation, actions or persistence) filter and print the values that repeat more than 10 times on RDD created on item (b). Note that these values should be printed in increasing order.

#### Displaying 500 random numbers in rdd

```
rdd.take(500)
```

#### Applying a Map and Reduce on the numbers and their frequencies

```
counts = rdd.map(lambda number: (number, 1)).reduceByKey(lambda a,b:a + b)
```

#### Filter Key Value pairs where value >= 10 and sort them in ascending order

```
values_that_appear_more_than_10_times = counts.sortBy(lambda keyVal: keyVal[1] , ascending = True) \
                                              .filter(lambda keyVal: keyVal[1] >= 10).collect()
values_that_appear_more_than_10_times
```
